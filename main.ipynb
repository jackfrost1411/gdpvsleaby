```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("final.csv")
df = df.rename(columns = {'Life expectancy at birth (years)':"LEABY"})
```

```python
#Bar Charts To Compare Average
plt.bar(df["Country"], df['GDP'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(df["Country"], df['LEABY'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Violin Plots To Compare Life Expectancy Distributions
fig = plt.subplots(figsize=(12, 7))

sns.violinplot(data=df, x='Country', y='LEABY',fontsize='large', fontweight='bold')
plt.savefig("violin.png",bbox_inches='tight')
```

```python
#Bar Plots Of GDP and Life Expectancy over time
```

```python
f, ax = plt.subplots(figsize=(10, 15)) 
ax = ax = sns.barplot(data=df,x="Country", y="GDP", hue='Year')
plt.xticks(rotation=90)
plt.ylabel("GDP in Trillions of U.S. Dollars")
```

```python
f, ax = plt.subplots(figsize=(10, 15)) 
ax = ax = sns.barplot(data=df,x="Country", y="LEABY", hue='Year')
plt.xticks(rotation=90)
plt.ylabel("Life expectancy at birth in years")
```

```python
#Scatter Plots of GDP and Life Expectancy Data
g = sns.FacetGrid(df, col='Year', hue='Country', col_wrap=4, size=2)
g = (g.map(plt.scatter, 'GDP', 'LEABY', edgecolor="w").add_legend())
#plt.savefig("facet.png",bbox_inches='tight')
```

```python
#Line Plots for LEABY
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, size=4)
g3 = (g3.map(plt.plot, "Year", "LEABY").add_legend())
#plt.savefig("facetline.png",bbox_inches='tight')
```

```python
#Line Plots for GDP
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, size=4)
g3 = (g3.map(plt.plot, "Year", "GDP").add_legend())
#plt.savefig("facetgdp.png",bbox_inches='tight')
```
